# Stats Demonstration

```python
import numpy as np
from matplotlib import pyplot as plt
```

## Generate some data

```python
mu, sigma = 23, 5
n = 50
x = np.random.normal(mu, sigma,n)
```

```python
counts, edges, bar = plt.hist(x, edgecolor='white')
tmp = plt.bar_label(bar)
```

# Point Estimation
## On this patch of data
### Calculate the Sample Mean

```python
mu_hat = sum(x)/n
print(mu_hat)
```

### Bias of this data 

```python
bias = mu-mu_hat
print(bias)
```

## Do Point Estimation on More Patches of Data
Say, let's generate 50 sets of data. For each set of data, calculate its sample mean and bias compared to $\mu$

```python
round = 50
bias = []
for i in range(round):
    xi = np.random.normal(mu, sigma, n)
    mu_hat_i = sum(xi)/n 
    bias_i = mu-mu_hat_i
    bias.append(bias_i)
```

```python
sum(bias)/n
```

```python
trial_n = np.arange(0,50)
```

```python
plt.plot(trial_n,bias, '.', label='bias on one dataset')
plt.plot(trial_n,[0]*n, '-', label='zero line')
plt.plot(trial_n, [sum(bias)/n]*n, '-', label='average of bias of average 50 sets of generated data')
plt.legend(loc='lower left')
```
